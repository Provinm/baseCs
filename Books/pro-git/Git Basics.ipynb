### git commit 

basic use:

git commit -m 'your commit'

skip `git add` step:

`git commit -a -m 'your commit '`



### git rm 

rm a file both on work directory and hard drive

`git rm test.txt -f`

rm a file only on work directory but keep it on hard drive

`git rm --cached test.txt`



### git mv

change a file name to another

`git mv file_a file_b`

### git log

look back to see what has happened on the project

`git log`


paras:

to see diff

`git log -p`

`git log -p -2`

to see status

`git log --stat`

to change default output

`git log --pretty=?`

? = [oneline, short, full, fuller, format, --graph]

please have a try above

`git log --pretty=format:"%h - %an, %ar : %s"`



para 'graph' 

`git log --pretty=format:"%h - %an, %ar : %s --graph`


to filter log history

`git log --since='2017-11-10'`

`git log --until='2017-11-10'`

`git log --author='zhouxin'`

`git log --grep='XXX'`



### Undoing Things

 commit 

`git commit --amend `

`git commit --amend -m 'changed text'`



### git reset

back to previous version, such as 'HEAD'

`git reset HEAD` 

unstage a file to former version

`git reset HEAD lala.py`




### Unmodifying a Modified File


diacard a file's modifies

`git checkout -- file.py`



### git remote 

`git remote`

`git remote -v`

remote address can be multiple, add one 

`git remote add origin2 http://...`

then you can fetch data from it

`git fetch origin2`



### git fetch

`git fetch <remote>` only get remote data

`git merge` merge remote data

and you can use `git pull ` to do the two steps in one command


### git push

`git push <remote> <branch>`

`git push` is short for `git push origin master`
