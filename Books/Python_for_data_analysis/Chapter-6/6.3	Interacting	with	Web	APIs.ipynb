## 6.3	Interacting	with	Web	APIs

```python
import numpy as np
import pandas as pd
import requests
```

```python
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
```

```python
resp = requests.get(url)
```

```python
resp
```

```python
data = resp.json()
```

```python
data[0]['title']
```

```python
issues = pd.DataFrame(data, columns='numbers title labels state'.split())
```
