## 4.7	Example:	Random	Walks

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import random
```

```python
position = 0
walk = [position]
```

```python
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)
```

```python
plt.plot(walk[:1000])
```

```python
plt.show()
```

```python
# using np

nsteps = 1000
draws = np.random.randint(0,2, size=nsteps)
```

```python
steps = np.where(draws>0, 1, -1)
```

```python
walk = steps.cumsum()
```

```python
walk.min()
```

```python
walk.max()
```

```python
plt.plot(walk)
```

```python
plt.show()
```

```python
# argmax()
(np.abs(walk) >= 10).argmax()
```

### Simulating	Many	Random	Walks	at	Once

```python
nwalks = 5000
nsteps = 1000

draws = np.random.randint(0,2, size=(nwalks, nsteps))
```

```python
steps = np.where(draws>0, 1, -1)
```

```python
walk = steps.cumsum(axis=1)
```

```python
walk
```

```python
walk.max()
```

```python
walk.min()
```

```python
hits30 = (np.abs(walk) >= 30).any(1)
```
