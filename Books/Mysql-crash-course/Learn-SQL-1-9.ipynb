### 4. Mysql 简单使用

#### 基本命令

连接本地数据库

`$ mysql -uroot -pxxxxxx`

显示目前所有数据库

```
$ show databases;


+--------------------+
| Database           |
+--------------------+
| information_schema |
| job                |
| lab                |
| mysql              |
| performance_schema |
| recruitment        |
| sys                |
| tools              |
+--------------------+

```

连接某一特定的数据库，显示所有的表

```linux

$ use job;
$ show tables;

+---------------+
| Tables_in_job |
+---------------+
| testitem      |
+---------------+

```



查看表内的列数据

```
$ show columns from testitem;

+--------+--------------+------+-----+---------+----------------+
| Field  | Type         | Null | Key | Default | Extra          |
+--------+--------------+------+-----+---------+----------------+
| id     | int(11)      | NO   | PRI | NULL    | auto_increment |
| field1 | varchar(255) | NO   |     | NULL    |                |
| field2 | int(11)      | NO   |     | NULL    |                |
+--------+--------------+------+-----+---------+----------------+


```



#### 查询命令

```
# 查询所有列
$ select * from testitem;

# 查询某一列
$ select id from testitem;

# 查询限制
$ select id from testitem limit 5;

# 类似切片的查询,从 1 开始， 查询一条
$ SELECT id FROM testitem LIMIT 1,1;
```
