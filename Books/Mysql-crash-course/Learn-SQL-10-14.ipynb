#### 13.1 GROUP BY 关键字


GROUP BY 关键字 为分组用 

```


mysql> SELECT port, COUNT(*) FROM myproxy_proxy GROUP BY port LIMIT 10;
+-------+----------+
| port  | COUNT(*) |
+-------+----------+
| 10000 |        1 |
| 10001 |        1 |
| 10066 |        1 |
| 10299 |        1 |
| 10412 |        1 |
| 10640 |        1 |
| 1080  |       43 |
| 1081  |        1 |
| 11345 |        1 |
| 11355 |        1 |
+-------+----------+

```

#### 13.2 HAVING 关键字


```

mysql> SELECT port, COUNT(*) FROM myproxy_proxy GROUP BY port 
    -> HAVING COUNT(*) > 10
    -> LIMIT 10;
+-------+----------+
| port  | COUNT(*) |
+-------+----------+
| 1080  |       43 |
| 1189  |       36 |
| 3128  |      297 |
| 33862 |       37 |
| 53281 |       92 |
| 63909 |       17 |
| 80    |      448 |
| 8000  |       16 |
| 808   |     3113 |
| 8080  |      566 |
+-------+----------+


```

#### 在上面的基础上排序

```

mysql> SELECT port, COUNT(*) FROM myproxy_proxy GROUP BY port HAVING COUNT(*) > 10 ORDER BY COUNT(*) LIMIT 10;
+-------+----------+
| port  | COUNT(*) |
+-------+----------+
| 81    |       13 |
| 9745  |       16 |
| 8000  |       16 |
| 63909 |       17 |
| 9001  |       17 |
| 8088  |       33 |
| 1189  |       36 |
| 33862 |       37 |
| 1080  |       43 |
| 8888  |       51 |
+-------+----------+

```
