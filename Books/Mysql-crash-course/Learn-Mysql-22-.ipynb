### 22.视图

视图是在真实的表的基础生成一个虚拟的表，比如某个表含有 5 个字段，我只希望开放其中 3 个字段给查询者，比如经常会使用到两个表的联接，每次写 sql 语句有些复杂等等，这些情况下都可以使用视图

- 重用 sql 语句
- 简化 sql 语句写法
- 保护表
等等


注意：

- 视图不包含任何数据，有点像数据映射
- 视图命名唯一
- 视图可以抽象为另一张表，所以可以嵌套
- ORDER BY 谨慎使用



#### 22.1 使用

- 使用 CREATE VIEW view1 AS SELECT ... 创建表
- SHOW CREATE VIEW viewname; 查看创建的表
- DROP VIEW viewname 删除表


举例： 从单词章节和词汇表中拿出 章节名和词汇 组成一个VIEW

```
# 创建view
mysql> CREATE VIEW cword AS
    -> SELECT chap, word
    -> FROM recite_chapter, recite_word
    -> WHERE recite_chapter.id = chapter_id;

# 使用 view
mysql> SELECT * 
    -> FROM cword
    -> WHERE chap =1 ;
+------+---------------+
| chap | word          |
+------+---------------+
|    1 | argument      |
|    1 | array         |
|    1 | attribute     |
|    1 | chapter       |
|    1 | class         |
|    1 | classes       |
|    1 | common        |
|    1 | create        |
|    1 | decorator     |
|    1 | default       |
|    1 | dictionary    |
|    1 | different     |
|    1 | discussion    |
|    1 | document      |
|    1 | documentation |
|    1 | example       |


```

其他用途：

- 使用视图重新格式化数据， 类似 .format
- 过滤数据
- 计算字段
- 更新

### 23.存储过程

存储过程可以理解为 Mysql 中的函数，床将一个存储过程，类似于创建一个查询或者计算或者其他的一个方法，使用时直接调用这个方法即可。

一个完整的过程

```
# 修改分割符
mysql> DELIMITER //

# 创建存储过程
mysql> CREATE PROCEDURE lab.check()
    -> BEGIN 
    ->   SELECT * FROM recite_word;
    -> END //

# 恢复分割符号
mysql> DELIMITER ;

# 调用存储过程
mysql> CALL lab.check()
    -> ;

```
