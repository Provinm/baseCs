#### 15.1 关系表

外键

所谓的联接，其实表与表之间没有任何物理链接，表是独立的
联接的原理是这个表中某个数值和另一个表中某个数值相等，查找便可以关联起来

比如：

设计一个单词表， 这个单词表有很多章， 每一章有很多单词

也叫 等值联接， 它基于两个表之间的相等测试。这种联结也称为内部联结。如 15.1 最后例子的结果相似，其具体的语法如下


```

mysql> SELECT chap,word 
    -> FROM recite_chapter INNER JOIN recite_word
    -> ON recite_chapter.id = chapter_id
    -> LIMIT 10;
+------+-----------+
| chap | word      |
+------+-----------+
|    1 | argument  |
|    1 | array     |
|    1 | attribute |
|    1 | chapter   |
|    1 | class     |
|    1 | classes   |
|    1 | common    |
|    1 | create    |
|    1 | decorator |
|    1 | default   |
+------+-----------+


```

单词表
包含 10 个章节，其中 9 个章节包含单词

使用内联接，则只会检索到包含单词的 9 个章节
而现在我们希望把不包含单词的章节也检索出来，这时就需要外联接。

外联接的关键词为 table1 LEFT/RIGHT OUTER JOIN table2  ON  conditions
其中 LEFT / RIGHT 分别表示外联表1还是表2

比如使用 chaper LEFT OUTER JOIN word ON conditions

表示检索 chapter, 即使 某些章节没有单词也拿出来，相反 RIGHT 表示检索所有单词，即使某些单词不属于任何章节。



### 17. UNION 组合查询

####  UNION 关键字

union 关键字的作用是合并多个 select 查询的结果

依然以单词表为例，我想查询第一章和第三章的单词，虽然这里可以使用 WHERE IN 语句，不过这次看看 UNION 怎么实现。

```


mysql> SELECT id, word, chapter_id 
    -> FROM recite_word
    -> WHERE chapter_id = 1
    -> UNION
    -> SELECT id, word, chapter_id
    -> FROM recite_word
    -> WHERE chapter_id = 10
    -> ;
+-----+---------------+------------+
| id  | word          | chapter_id |
+-----+---------------+------------+
|  27 | argument      |          1 |
|  12 | array         |          1 |
|  28 | attribute     |          1 |
|  17 | chapter       |          1 |
|   3 | class         |          1 |
|  30 | classes       |          1 |
|  47 | common        |          1 |
# ...


```

#### UNION  ALL 关键字

在某些情况下两个或多个 SELECT 语句会有重复的条目发生，为了解决这一问题， UNION ALL 关键字应运而生。

在原语句上直接将 UNION 改为 UNION ALL 即可。

### 18. 全文本搜索


- innoDB 引擎并不支持全文搜索
- MYISAM 引擎支持全文搜索

- LIKE 和 RE 可以做到类似全文搜索的功能，但性能上不如 MATCH AGAINST

#### 18.2 全文搜索示例

使用全文搜索需要在建表的时候使用 FULLTEXT 注明，类似加上索引。

```
mysql> CREATE TABLE sql_articles
    -> ( 
    -> id  int NOT NULL AUTO_INCREMENT,
    -> text text NULL,
    -> PRIMARY KEY(id),
    -> FULLTEXT(text)
    -> )ENGINE=MyISAM;

```


使用方法

这里遇到一个小问题，使用教程中常规的语句并不能缩影出东西
加了布尔模式后正常

```
# 无匹配
mysql> SELECT text  FROM sql_articles WHERE Match(text) Against('+conditional');
mysql> SELECT text  FROM sql_articles WHERE Match(text) Against('conditional');


# 正常
mysql> SELECT text  FROM sql_articles WHERE Match(text) Against('+conditional' IN BOOLEAN MODE);
mysql> SELECT text  FROM sql_articles WHERE Match(text) Against('conditional' IN BOOLEAN MODE);


```



#### 常见的操作符

布尔操作符 及 说明
 
- + 包含,词必须存在
- - 排除,词必须不出现
- \> 包含,而且增加等级值
- < 包含,且减少等级值
- () 把词组成子表达式(允许这些子表达式作为一个组被包含、排除、排列等)
- ~ 取消一个词的排序值
- * 词尾的通配符
- "" 定义一个短语(与单个词的列表不一样,它匹配整个短语以便包含或排除这个短语)
