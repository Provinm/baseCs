## HTTP协议知识

### 简介

- http 为 超文本传输协议
- http 基于 tcp/ip 协议传递数据（文本/图片/视频）
- http 协议工作于 客户端-服务端 
- http 协议默认端口 80
- 特点一： 无连接， 客户端--服务端 仅保持一次连接 请求 --> 响应
- 特点二： 媒体独立， 客户端与服务端各自有数据处理方式，互补干扰
- 特点三： 无状态， 请求与请求之间不会存在联系

http 结构

web browser --- http server --- cgi --- database



### HTTP结构

- http 客户端为一个应用程序 浏览器/脚本 等
- http 服务端也为一个应用程序  apache / nginx , 接收客服端请求发送响应数据
- URL 为统一资源标识符， 用于建立连接传输资源


#### 请求的消息包含

- 请求行
- 请求头部
- 空行
- 请求数据




#### 响应的消息

- 状态行
- 消息报头
- 空行
- 正文



### HTTP 请求方法

请求方法：

- GET   请求指定页面，返回数据
- POST  提交数据
- HEAD  返回报头
- PUT   向服务器传输内容取代指定的内容
- DELETE 删除制定的内容
- OPTIONS 查看服务器性能
- CONNECT 代理服务器
- TRACE 测试诊断服务器

#### HTTP 状态码

- 1** 收到请求，需要操作者继续操作
- 2** 成功
- 3** 重定向
- 4** 客户端错误
- 5** 服务器错误
